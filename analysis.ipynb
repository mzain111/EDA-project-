## 6. Key Insights

```python
print('=== KEY INSIGHTS FROM ETH DATASET ===')
print(f"Date Range: {df['date'].min().date()} to {df['date'].max().date()}")
print(f"Total Trading Days: {len(df)}")
print(f"Average Price: ${df['price'].mean():,.2f}")
print(f"Highest Price: ${df['price'].max():,.2f}")
print(f"Lowest Price:  ${df['price'].min():,.2f}")
print(f"Total Return:  {((df['price'].iloc[-1]-df['price'].iloc[0])/df['price'].iloc[0]*100):.2f}%")
print(f"Bullish Days:  {(df['price_direction']=='Bullish').sum()}")
print(f"Bearish Days:  {(df['price_direction']=='Bearish').sum()}")
print(f"Price-Volume Correlation: {df['price'].corr(df['total_volume']):.4f}")
print(f"Price-MarketCap Corr:    {df['price'].corr(df['market_cap']):.4f}")
```
